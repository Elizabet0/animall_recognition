```python
import os
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
```

```python
from torch.utils.data import random_split
import torch
```

```python
data_path=Path("../data/raw-img")

dataset = ImageFolder(data_path) # обычный датасетик
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)), # всё в один размер
    transforms.ToTensor() #тензор
])

#готовый датасет
dataset_t = ImageFolder(str(data_path), transform=transform) # Теперь элемент - тензор вида [3, 224, 224], канал ширины высота

#делим в отношение 70/15/15
train_size=int(0.7*(len(dataset_t)))
val_size=int(0.15*(len(dataset_t)))
test_size=len(dataset_t)- train_size -val_size   

#делим на выбоки
train_ds, val_ds, test_ds = random_split(dataset_t,
    [train_size, val_size, test_size],
    generator=torch.Generator().manual_seed(42))
```

```python
torch.save(train_ds, '../data/processed/train_ds.pt')
torch.save(val_ds, '../data/processed/val_ds.pt')
torch.save(test_ds, '../data/processed/test_ds.pt')
```
