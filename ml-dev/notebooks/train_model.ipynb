```python
import torch.nn as nn
import torch.nn.functional as F

class TinyModel(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        # Просто выпрямляем картинку и делаем линейный слой
        self.fc = nn.Linear(3 * 224 * 224, num_classes)  # 3*224*224 -> 10 классов
    
    def forward(self, x):
        # x shape: [batch, 3, 224, 224]
        x = x.view(x.size(0), -1)  # Выпрямляем в [batch, 3*224*224]
        x = self.fc(x)  # Прямой проход
        return x

# Создаем модель
model = TinyModel(num_classes=10)
```
